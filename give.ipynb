```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  # Import LinearRegression model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error 
import math
excel_file1 = 'C:\\Users\\satya\\OneDrive\\Desktop\\cleans.xlsx'
df = pd.read_excel(excel_file1)
csv_file = 'stry.csv'  
df.to_csv('stry.csv', index=False)
df['Orde rDate'] = pd.to_datetime(df['Order Date'], format='%m-%d-%Y')
df2019 = df[df['Order Date'].dt.year == 2019]
df2020 = df[df['Order Date'].dt.year == 2020]
df2020.drop('Profit',axis=1,inplace=True)
from sklearn.linear_model import Lasso
cd=["Sales","Quantity"]
X=df2019[cd].values
y=df2019["Profit"].values
names=cd
lasso=Lasso(alpha=0.1)
lasso_coef = lasso.fit(X, y).coef_
plt.bar(names, lasso_coef)
plt.xlabel('Features')
plt.ylabel('Coefficient')
plt.title('Lasso Coefficients')
plt.xticks(rotation=45)
plt.show()
# Splitting data into features (X) and target variable (y)
X = df2019['Quantity']  # Select relevant features
y = df2019['Profit']  
X_reshaped=X.values.reshape(-1,1)
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
# Define a range of k values to experiment with
k_values = [3, 5, 7, 10, 12, 15]  # Adjust this range as needed
for k in k_values:
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    reg = LinearRegression()
    cv_results = cross_val_score(reg, X_reshaped, y, cv=kf)
    mean_score = cv_results.mean()  # Calculate the mean score
    print(f"k = {k}, Mean Score: {mean_score}")
    
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
X_train = X_train.values.reshape(-1, 1)  # Reshape X_train if it's a single feature
y_train=y_train.values.reshape(-1,1)
X_test_reshaped = X_test.values.reshape(-1, 1)
model = LinearRegression()  # Instantiate a Linear Regression model
model.fit(X_train, y_train)  # Train the model on the training data
# Model Evaluation
y_pred = model.predict(X_test_reshaped)  # Make predictions on the test set
mse = mean_squared_error(y_test, y_pred)  # Calculate Mean Squared Error
print(f"Mean Squared Error: {mse}")
print(math.sqrt(mse))
# Making Predictions for the Next Year (Example: Creating new features for the next year)
# Assuming 'NextYearData' contains features for the next year
NextYearData =df2020[['Quantity']] 
NextYearData_reshaped = NextYearData.values.reshape(-1, 1)
# Use the trained model to predict counts for the next year
next_year_predictions = model.predict(NextYearData)
print("Predictions for the next year:")
print(next_year_predictions)
df2019['YearMonth'] = df2019['Order Date'].dt.to_period('M')
monthly_sales = df2019.groupby('YearMonth')['Profit'].sum()
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Profits')
plt.xlabel('Month')
plt.ylabel('Total Profits')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
df2020['new1']=next_year_predictions
df2020['YearMonth'] = df2020['Order Date'].dt.to_period('M')
monthly_sales = df2020.groupby('YearMonth')['new1'].sum()
# Plotting the line graph
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='red')
plt.title('Monthly Profits')
plt.xlabel('Month')
plt.ylabel('Total Profits')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
