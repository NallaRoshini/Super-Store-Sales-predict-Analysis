```python
import pandas as pd
from sklearn.linear_model import Lasso
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression  # Import LinearRegression model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error 
import numpy as np
import time
import math
excel_file1 = 'C:\\Users\\satya\\OneDrive\\Desktop\\cleans.xlsx'
df = pd.read_excel(excel_file1)
csv_file = 'stry.csv'  
df.to_csv('stry.csv', index=False)

payment_method = {'COD': '1', 'Online': '2', 'Cards': '3'}

# Transform 'Payment Mode' column using payment_method dictionary
df['Payment Mode'] = df['Payment Mode'].map(payment_method).fillna(df['Payment Mode'])
df['Payment Mode'] = pd.to_numeric(df['Payment Mode'])

# Label Encoding for categorical columns
label_encoder = LabelEncoder()

# Transform 'Segment', 'Region', 'Ship Mode' columns
df['Segment'] = label_encoder.fit_transform(df['Segment'])
df['Region'] = label_encoder.fit_transform(df['Region'])
df['Ship Mode'] = label_encoder.fit_transform(df['Ship Mode'])

df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m-%d-%Y')

df2019 = df[df['Order Date'].dt.year == 2019]
df2020 = df[df['Order Date'].dt.year == 2020]
cd = ['Region', 'Segment', 'Ship Mode', 'Quantity','Payment Mode','Profit']
X = df2019[cd].values
y = df2019["Sales"].values
names = cd
X1 = df2019[cd]  
y1= df2019["Payment Mode"].values

a1=np.arange(1.0,100.0,1.0)
scores=[]
for alpha in a1:
    l1=Lasso(alpha=alpha) 
    l1.fit(X1,y1)
    lp=l1.predict(X1)
    scores.append(l1.score(X1,y1))
print(scores.index(max(scores)))
lasso = Lasso(alpha=1.0)
lasso_coef = lasso.fit(X, y).coef_
plt.figure(figsize=(8, 6))
plt.bar(names, lasso_coef)
plt.xticks(rotation=45)
plt.show()
print(lasso_coef)

##########
df2020.drop('Sales',axis=1,inplace=True)
# Splitting data into features (X) and target variable (y)
X = df2019['Quantity']  # Select relevant features
y = df2019['Sales'] 
X_reshaped=X.values.reshape(-1,1) # 
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
# Define a range of k values to experiment with
k_values = [3, 5, 7, 10, 12, 15]  # Adjust this range as needed
for k in k_values:
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    reg = LinearRegression()
    cv_results = cross_val_score(reg, X_reshaped, y, cv=kf)
    mean_score = cv_results.mean()  # Calculate the mean score
    print(f"k = {k}, Mean Score: {mean_score}")

# Apply LabelEncoder to your target variable and replace values in y_train
start_time_training = time.time()
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)
X_train = X_train.values.reshape(-1, 1)  # Reshape X_train if it's a single feature
y_train=y_train.values.reshape(-1,1)
X_test_reshaped = X_test.values.reshape(-1, 1)
model = LinearRegression()  # Instantiate a Linear Regression model
model.fit(X_train, y_train)  # Train the model on the training data
end_time_training = time.time()
training_time = end_time_training - start_time_training
# Model Evaluation
start_time_predictions = time.time()
y_pred = model.predict(X_test_reshaped) 
end_time_predictions = time.time()
prediction_time = end_time_predictions - start_time_predictions # Make predictions on the test set
mse = mean_squared_error(y_test, y_pred)  # Calculate Mean Squared Error
print(f"Mean Squared Error: {mse}")
print(math.sqrt(mse))
print(f"Time taken for training: {training_time} seconds")
print(f"Time taken for predictions: {prediction_time} seconds")
# Making Predictions for the Next Year (Example: Creating new features for the next year)
# Assuming 'NextYearData' contains features for the next year
NextYearData =df2020['Quantity'] 
NextYearData_reshaped = NextYearData.values.reshape(-1, 1)
# Use the trained model to predict counts for the next year
next_year_predictions = model.predict(NextYearData_reshaped)
print("Predictions for the next year:")
print(next_year_predictions)

df2019['YearMonth'] = df2019['Order Date'].dt.to_period('M')

# Aggregate sales by month
monthly_sales = df2019.groupby('YearMonth')['Sales'].sum()

# Plotting the line graph
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='blue')

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
df2020['new1']=next_year_predictions 
df2020['YearMonth'] = df2020['Order Date'].dt.to_period('M')

# Aggregate sales by month
monthly_sales = df2020.groupby('YearMonth')['new1'].sum()

# Plotting the line graph
plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='line', marker='o', color='red')

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
x_values_scatter = df2019['Sales'].values
y_values_scatter = df2019['Quantity'].values
x_values_line =next_year_predictions 
y_values_line = df2020['Quantity'].values
# Create a figure and axis
fig, ax = plt.subplots()

# Plotting scatter plot
ax.scatter(x_values_scatter, y_values_scatter, label='Scatter', color='red')

# Plotting line plot
ax.plot(x_values_line, y_values_line, label='Line', color='blue')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter and Line Plot')

# Show legend
ax.legend()

# Show the plot
plt.show()
```
