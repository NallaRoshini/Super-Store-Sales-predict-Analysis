```python
import pandas as pd
import numpy as np

excel_file ='C:\\Users\\nalla_enejoi3\\OneDrive\Desktop\\fda project\\SuperStore.xlsx' 
df = pd.read_excel(excel_file)
print(df)

#DATA CLEANING


#to print 1st few rows
print(df.head())
#to print last few rows
print(df.tail())
```

```python
#Removing if any duplicates are present
print(df.duplicated())
duplicates = df[df.duplicated()]
print(duplicates)
#i.e. no duplicated rows are present in the dataset(empty dataframe)
print(duplicates.value_counts())
```

```python
# To print the first few elements of the 'Region' column
print(df['Region'].head())
# To print the last few elements of the 'Region' column
print(df['Region'].tail())


# Mapping dictionary for abbreviation to full name
region_mapping = {"E": "East", "S": "South", "C": "Central", "W": "West"}

# Iterate through each row in the 'Region' column
for i in range(len(df)):
    if df['Region'][i] in region_mapping:
        df.at[i, 'Region'] = region_mapping[df['Region'][i]]

# To print the first few elements of the updated 'Region' column
print(df['Region'].head())
# To print the last few elements of the updated 'Region' column
print(df['Region'].tail())

#to print updated 'Region' column
print(df['Region'])
```

```python
# To print the first few elements of the 'Quantity' column
print(df['Quantity'].head())
# To print the first few elements of the 'Quantity' column
print(df['Quantity'].tail())

quantity = ["One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten"]

# Iterate through each row in the 'Quantity' column
for i in range(len(df)):
    if df['Quantity'][i] in quantity:
        # Find the index of the quantity in the list and add 1 to get the numeric value
        df.at[i, 'Quantity'] = str(quantity.index(df['Quantity'][i]) + 1)

# To print the first few elements of the updated 'Quantity' column
print(df['Quantity'].head())
# To print the first few elements of the updated 'Quantity' column
print(df['Quantity'].tail())

# coverting 'Quantity' values to integer datatype
df['Quantity'] = df['Quantity'].astype(int)

```

```python
#To print the first few elements of the 'Sales' column
print(df['Sales'].head())
#To print the last few elements of the 'Sales' column
print(df['Sales'].tail())

# checking if NA values are present in 'Sales' column or not.
missing_values = df['Sales'].isna()
print(missing_values)
# counting total no of NAs
missing_sales_sum = df['Sales'].isna().sum()
print(missing_sales_sum)

# filling NA values with mean of the 'Sales' column
mean_sales = df['Sales'].mean(skipna=True)
df['Sales'].fillna(mean_sales, inplace=True)

print(df['Sales'].head())
print(df['Sales'].tail())
```

```python
print(df['Order Date'])
any_missing = df.isna().any().any()
print(any_missing)
number_of_na = df.isna().values.sum()

if number_of_na > 0:
    df = df.dropna()
    
any_missing_after_removal = df.isna().any().any()
print(any_missing_after_removal)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m-%d-%Y')
# Sort the DataFrame by the 'Order.Date' column in ascending order
df.sort_values(by='Order Date', inplace=True)
print(df['Order Date'])


```

```python
# encoding categorical variable:Payment mode
category=df['Payment Mode'].value_counts()
print(category)

payment_method={'COD':'1','Online':'2','Cards':'3'}

# Iterate through each row in the 'Region' column
for i in range(len(df)):
    if df['Payment Mode'][i] in payment_method:
        df.at[i, 'Payment Mode'] = payment_method[df['Payment Mode'][i]]

df['Payment Mode'] = pd.to_numeric(df['Payment Mode'])

print(df['Payment Mode'])
```
